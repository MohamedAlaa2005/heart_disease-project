```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
heart_disease=pd.read_csv(r"Data/heart_disease_scaled.csv")
```

```python
x=heart_disease.iloc[:,:-1]
y=heart_disease['num']
```

```python
pca = PCA()
x_pca=pca.fit_transform(x)
# draw variance_ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python

num_classes = y.nunique()
palette = sns.color_palette("RdYlGn_r", num_classes)
# PCA Scatter Plot (PC1 vs PC2)
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=x_pca[:, 0], 
    y=x_pca[:, 1], 
    hue=y, 
    palette=palette, 
    alpha=0.7,
    legend='full'
)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot - Heart Disease Classes')
plt.legend(title='Heart Disease Level')
plt.grid(True)
plt.show()
```

```python
x_pca = pd.DataFrame(x_pca[:, :10], columns=[f'pc{i+1}' for i in range(10)])
x_pca
```

```python
#concut traget to pca compond
y=pd.DataFrame(y)
heart_disease_pca=pd.concat([x_pca,y],axis=1)
```

```python
# save data
heart_disease_pca.to_csv("Data\heart_disease_pca.csv", index=False)
```
